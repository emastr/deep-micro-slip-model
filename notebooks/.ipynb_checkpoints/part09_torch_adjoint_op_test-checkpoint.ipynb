## Explaining weird patterns.

The kernel matrices exhibit non-local behaviour, why is that? Let's investigate the kernel. For a circle with parametrisation

$ \gamma(t) = e^{it}, \qquad t \in [0, 2\pi]$

the operator can be described as 

$v(t) = \int_\gamma k(t,s)\rho(s)\mathrm{d}s + \int_\gamma \overline{k}(t,s)\overline{\rho}(s)\mathrm{d}s$

where

$k(t,s) = \mathrm{Im}\left\{\frac{\gamma'(t)}{\gamma(t)-\gamma(s)}\right\} = \mathrm{Im}\left\{\frac{ie^{it}}{e^{it}-e^{is}}\right\}=\frac{\mathrm{Im}\left\{ie^{it}(e^{-it}-e^{-is})\right\}}{(e^{it}-e^{is})(e^{-it}-e^{-is})}=\frac{\mathrm{Re}\left\{1-e^{i(t-s)}\right\}}{2-e^{i(t-s)}-e^{i(s-t)}}=\frac{1-cos(t-s)}{2(1-cos(t-s))}=\frac{1}{2}$

and

$\overline{k}(t,s) = \frac{\mathrm{Im}\left\{\gamma'(t)(\overline{\gamma}(t)-\overline{\gamma}(s))\right\}}{(\overline{\gamma}(t)-\overline{\gamma}(s))^2} = \frac{\mathrm{Im}\left\{ie^{it}(e^{-it}-e^{-is})\right\}}{(e^ {-it}-e^{-is})^2} = \frac{\mathrm{Re}\left\{1-e^{i(t-s)})\right\}}{e^ {-2it}-2e^{-i(t+s)}+e^{-2is}}=\frac{1-\cos(t-s)}{e^{-i(t+s)}(e^{i(s-t)}-2+e^{i(t-s)})}=-\frac{1}{2}e^{i(t+s)}$

Hence, the kernel operators of a circular domain exhibit non-local dependencies, which would limit the accuracy of a naive cnn design.

```python
import matplotlib.pyplot as plt
from matplotlib import image as im
import numpy as np

x = im.imread("/home/emastr/phd/picture/Photo1.jpeg")[:,:,0].astype(float)/255
y = im.imread("/home/emastr/phd/picture/Photo2.jpeg")[:,:,0].astype(float)/255
print(x.shape)

#plt.imshow(x[:,:,2], cmap="Greys_r")
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(x)
plt.subplot(132)
plt.imshow(y)
plt.subplot(133)
plt.imshow(y-x)

#x = x[::50,::50]
#y = y[::50,::50]


def transfer(inp, phi):
    out = inp.copy()
    out[1:, :] += phi[:-1, :] * (inp[1:, :] - inp[:-1, :])
    #out[:-1, :] += phi[1:, :] * (inp[:-1, :] - inp[1:, :])
    out[:, 1:] += phi[:, :-1] * (inp[:, 1:] - inp[:, :-1])
    #out[:, :-1] += phi[:, 1:] * (inp[:, :-1] - inp[:, 1:])
    return out
    

def grad(inp, out, phi):
    # Precompute 
    M, N = inp.shape[0], inp.shape[1]
    ds = [[1,0],[0,1]]#,[-1,0],[0,-1]]
    inrange = lambda idx: (0 <= idx[0]) & (idx[0] < M) & (0 <= idx[1]) & (idx[1] < N)
    
    # Allocate
    gradL = np.zeros_like(phi)
    
    # Gradient of input
    gInp0 = inp[1:, :] - inp[:-1, :]
    gInp1 = inp[:, 1:] - inp[:, :-1]
    
    # Error
    tinp = transfer(inp, phi)
    err = (tinp - out)
    
    gradL[:-1,:-1] = 2 * (gInp0[] + gInp1[]) * err[:-1,:-1]
    gradL[1:-1,1:-1] = gradL[1:-1,1:-1] - 2 * (gInp0[1:,1:] * err[1:,1:] + gInp1[1:,1:] * err[])
    
    # Loop through image
    for m in range(M):
        for n in range(N):
            Nmn = [[m + d[0], n + d[1]] for d in ds]
            Nmn = [idx for idx in Nmn if inrange(idx)]
            
            # Gradient of image
            dPmn = [inp[m,n] - inp[idx[0], idx[1]] for idx in Nmn]
            dPsum = sum(dPmn)
            
            # Current error 
            
            
            # Gradient of mass transfer strategy
            gradL[m,n] = 2 * dPsum
    return gradL
            

plt.figure()
#gL = grad(x, y, np.zeros_like(x))
out = x
for i in range(6):
    out = transfer(out, out)
plt.imshow(out)
```

```python
import torch as t
from torch.autograd import grad
import matplotlib.pyplot as plt
import matplotlib.image as im

inp = im.imread("/home/emastr/phd/picture/Photo1.jpeg")[:,:,0].astype(float)/255
out = im.imread("/home/emastr/phd/picture/Photo2.jpeg")[:,:,0].astype(float)/255


N = 2
inp = inp[::N,::N]
out = out[::N,::N]


y1 = t.from_numpy(inp)
y2 = t.from_numpy(out)



def transfer(x, y):
    out = y.clone()
    gy0 = y[1:, :] - y[:-1, :]
    gy1 = y[:, 1:] - y[:, :-1]
    out[1:, :] += gy0 * x[:-1, :]
    out[:, 1:] += gy1 * x[:, :-1]
    # Transfer mass from y onto itself in the direction of the gradient, using masses x
    return out
    
def mse(y1, y2):
    # Mean square error of y and y2
    return ((y1 - y2)**2).mean()

def objective(x, y1, y2):
    return mse(transfer(x, y1), y2)


x = t.ones_like(y1)
x.requires_grad_()


loss = []
iters = []
J = 20
for j in range(J):
    print(f"{j}/{J} iterations")
    for i in range(1000):
        L = objective(x, y1, y2)
        gxL = grad(L, x)[0]
        x = x - 200 * gxL
        loss.append(L.item())
    iters.append(y1.clone())
    y1 = transfer(x, y1)
```
