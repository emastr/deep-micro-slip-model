## Sensitivity Analysis

Suppose u satisfies

$$
    \nu\Delta u - \nabla p = f, \quad \text{in} \quad \Omega
$$

$$
\nabla \cdot u = 0\quad \text{in}\quad\Omega
$$

with boundary condition in tangental component $u_\tau$ and normal component $u_n$:

$$
u_\tau + \alpha\partial_n u_\tau = \beta,\quad \text{on}\quad \partial\Omega
$$

$$
u_n = 0\quad \text{on} \quad \partial\Omega
$$

Then, we can bound $||u||^2$ as follows. First, use the Point-Care inequality:

$$
    ||u||^2 \leq V_\Omega ||\nabla u||^2 = V_\Omega \int_\Omega( \nabla \cdot (u\cdot \nabla u) - u\cdot \Delta u)\mathrm{d}x
$$

Next, use the pde: $u\cdot \Delta u = \nu^{-1}u \cdot f + \nu^{-1}u\cdot \nabla p = \nu^{-1}u\cdot f + \nu^{-1}\nabla \cdot (u p)$:

$$
    ||u||^2 \leq V_\Omega \left(\int_{\partial\Omega} (u\cdot \partial_n u - \nu^{-1} u_n p) \mathrm{d}s - \nu^{-1}\int_\Omega u\cdot f \mathrm{d}x \right),
$$

Whereby the boundary conditions give us

$$
    ||u||^2 \leq V_\Omega \left(\int_{\partial\Omega}(u_\tau \frac{\beta}{\alpha} - \frac{1}{\alpha}|u_\tau|^2)\mathrm{d}s - \nu^{-1}\int_\Omega u\cdot f\mathrm{d}x\right).
$$

The right expression can be bounded from above by using the trace inequality: $||u||_{\partial\Omega} \leq C||u||$, so we obtain

$$
    ||u||^2 \leq V_\Omega \left(C\alpha_{inf}^{-1}||u||\cdot ||\beta||_{\partial\Omega} + C^2\alpha_{inf}^{-1}||u||^2 + \nu^{-1}||f||\cdot ||u||\right)
$$

Rearranging and assuming that $1 - V_\Omega C^2\alpha_{inf}^{-1} \geq 0$, the final bound for $||u||$ is

$$
    ||u||\leq \frac{V_{\Omega}(C\alpha_{inf}^{-1}||\beta||_{\partial\Omega} + \nu^{-1}||f||)}{1 - V_\Omega C^2\alpha_{inf}^{-1}}
$$


### Slip Coefficient Sensitivity

Suppose two solutions $u, v$ satisfy the above with different slip coefficients $\alpha_u, \alpha_v$ and $\beta=0$. Then, the difference $w=u-v$ satisfies the same equations, with $f=0$ and the following at the boundary:
$$
    w + \alpha_u w = (\alpha_v-\alpha_u)v\alpha_v^{-1}.
$$

Thus, the above bound gives

$$
    ||w||\leq \frac{V_{\Omega}C\alpha_{inf}^{-1}}{1 - V_\Omega C^2\alpha_{inf}^{-1}}||(\alpha_v-\alpha_u)\alpha_v^{-1}v||_{\partial\Omega}
$$

Suppose regularity such that $v$ is bounded on the boundary. We then obtain the following:
$$
    ||w||\leq \frac{V_{\Omega}C\alpha_{inf}^{-2}v_{sup}}{1 - V_\Omega C^2\alpha_{inf}^{-1}}||\alpha_v-\alpha_u||_{\partial\Omega} = \tilde{C}||\alpha_v-\alpha_u||_{\partial\Omega},
$$

where $\tilde{C}$ depends on $\alpha_{inf}$, $\Omega$, and $v_{sup}$. We make the assumption of a uniform bound on $v$, because bounding $||w||$ with the $2-norm$ allows us to use the Parseval identity, and work in a Hilbert space. Suppose that 

$$
    \alpha_u = \phi \circledast \alpha_v,
$$

where $\phi$ is a filter, in our case the square bump function with width $a$, so that $\hat{\phi}_n = \int_{-a}^a \frac{1}{2a} e^{-i\pi n x}\mathrm{d}x = \frac{e^{-i\pi n a} - e^{i\pi n a}}{-2i\pi a n} = \frac{\sin (\pi n a)}{\pi n a}$. Thus, we have

$$
    ||\alpha_v - \phi \circledast\alpha_v||_{\partial\Omega} = \sum_{n=-\infty}^\infty (1-\hat{\phi}_n)^2|\alpha_n|^2 
$$

Since $\alpha$ is $\epsilon$ scale, suppose $\epsilon > 1/m$ for some $m$ and assume $\alpha_n = 0$ for all $n > m$. Then,

$$
    ||\alpha_v - \phi \circledast\alpha_v||_{\partial\Omega} = \sum_{n=-m}^m (1-\hat{\phi}_n)^2|\alpha_n|^2
$$

### Standard multi scale assumption

Assume $\alpha(x) = \tilde\alpha(x) + \epsilon v(x/\epsilon)$, where $\tilde\alpha$ and $v$ are band-limited functions, such that $\hat\alpha_n, \hat v_n = 0$ for all $|n|\geq N$, some $N$ and $||\tilde\alpha||=||v||=1$. Then, if $\epsilon = 1/m$, we have

$$
    \phi \circledast\alpha = (\phi\circledast \tilde \alpha)(x) + (\epsilon \phi_\epsilon \circledast v)(x/\epsilon),\qquad ( \phi\circledast \alpha \;, \; e^{inx}) = \hat\phi_n\hat{\tilde\alpha}_n + \tfrac{1}{m}\hat\phi_{nm} \hat v_n.
$$

where $\phi_\epsilon(x) = \epsilon\phi(\epsilon x)$ is a stretched version of $\phi$. Now suppose we pick a sampling rate of $k > N$, meaning $\phi\circledast\tilde\alpha$ is fully resolved. Then,

$$
    ||\alpha - \phi\circledast \alpha|| \leq ||(\delta - \phi)\circledast \tilde\alpha|| + \epsilon||(\delta - \phi_\epsilon)\circledast v||
$$
