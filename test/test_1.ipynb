```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: 1/(np.exp(-x) + 1)
f_inv = lambda y: -np.log(1/y - 1)
df = lambda x: np.exp(-x)/(np.exp(-x) + 1)**2
x = np.linspace(-10,10,100)


plt.plot(x, f(x))
plt.plot(x, df(x))


plt.figure()
x = np.linspace(0, 1, 100)
plt.plot(x, f_inv(x))


X = np.random.randn(100000)
Y = f_inv(X)
plt.hist(Y, bins=100)
print("hej")
```
